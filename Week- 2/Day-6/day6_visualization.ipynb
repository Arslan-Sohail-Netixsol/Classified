# Essential EDA Charts
Building the 6 chart types that cover 90% of real EDA needs using the Retail Sales Dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Retail Sales Dataset
df = pd.read_csv(r'D:\data sets\reatil adta\retail_sales_dataset.csv')

# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

sns.set_theme(style="whitegrid")
df.head()
```

## 1. Line Chart
**What this tells us:** Shows the overall trend and fluctuations of total sales amounts over time.

```python
# Group by Date to get daily total sales
daily_sales = df.groupby('Date')['Total Amount'].sum().reset_index()

plt.figure(figsize=(12, 5))
sns.lineplot(data=daily_sales, x='Date', y='Total Amount')
plt.title('Daily Total Sales Over Time')
plt.xticks(rotation=45)
plt.show()
```

## 2. Bar Chart
**What this tells us:** Compares the aggregate total sales generated by each distinct product category.

```python
category_sales = df.groupby('Product Category')['Total Amount'].sum().reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(data=category_sales, x='Product Category', y='Total Amount')
plt.title('Total Sales by Product Category')
plt.show()
```

## 3. Histogram
**What this tells us:** Reveals the age distribution of customers, highlighting which age groups make the most purchases.

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='Age', bins=20, kde=True)
plt.title('Distribution of Customer Age')
plt.show()
```

## 4. Boxplot
**What this tells us:** Displays the spread, central tendency, and potential outliers in total transaction amounts across different product categories.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Product Category', y='Total Amount')
plt.title('Total Amount Spread by Product Category')
plt.show()
```

## 5. Scatter Plot
**What this tells us:** Explores the relationship and potential patterns between a customer's age and the amount they spend.

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Age', y='Total Amount', hue='Gender', alpha=0.7)
plt.title('Total Amount vs Customer Age')
plt.show()
```

## 6. Heatmap
**What this tells us:** Highlights strong positive or negative correlations between numerical variables (like Quantity and Total Amount) at a glance.

```python
plt.figure(figsize=(8, 6))
# Select only numerical columns for correlation
num_df = df[['Age', 'Quantity', 'Price per Unit', 'Total Amount']]
corr = num_df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()
```
