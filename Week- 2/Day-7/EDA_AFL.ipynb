# Exploratory Data Analysis (EDA) on AFL Dataset
This notebook presents the EDA of the merged AFL player datasets, answering specific business questions with visualizations, observations, and insights.

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Load dataset
df = pd.read_csv('merged_players.csv')
df.head()
```

## 1. Which teams have played the most games?
**Business Question:** Which AFL teams have the highest total games played by their players historically?

**Observations:**
- Collingwood, Carlton, and Essendon are among the top teams with the highest cumulative player games.
- The top teams show a relatively close number of total games, reflecting their long-standing presence in the league.

**Business Insight:** Teams with the highest total games played likely possess a richer historical dataset and larger fan bases, making them prime candidates for historical merchandise campaigns or legacy-focused marketing.

```
plt.figure(figsize=(12, 6))
team_games = df.groupby('team')['games_played'].sum().sort_values(ascending=False).head(15)
sns.barplot(x=team_games.values, y=team_games.index, palette='viridis')
plt.title('Top 15 Teams by Total Player Games Played')
plt.xlabel('Total Games Played')
plt.ylabel('Team')
plt.show()
```

## 2. What is the distribution of player ages (last_age)?
**Business Question:** What is the typical age range at which AFL players finish their careers or are last recorded?

**Observations:**
- The distribution is slightly right-skewed, peaking around the late 20s to early 30s.
- A significant drop-off occurs after age 32, with very few players remaining active past 35.

**Business Insight:** Knowing the typical retirement age allows clubs to better plan their list management and talent pipelines, focusing on drafting and developing younger talent as veterans approach the 30-year mark.

```
plt.figure(figsize=(10, 6))
sns.histplot(df['last_age'].dropna(), bins=30, kde=True, color='skyblue')
plt.title('Distribution of Player Ages (last_age)')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

## 3. Which player positions/teams have the highest number of players?
**Business Question:** Which historical team affiliations have the highest representation of players in the dataset?

**Observations:**
- The top individual player_teams are generally long-established clubs like Collingwood, Carlton, and Essendon.
- The distribution of player counts among the top teams is quite gradual, showing consistent squad sizes across major teams historically.

**Business Insight:** Clubs with the highest historical player turnover or larger historical squads might require deeper dive analysis into player retention strategies versus those with fewer unique players.

```
plt.figure(figsize=(12, 6))
team_players = df['player_teams'].value_counts().head(15)
sns.barplot(x=team_players.values, y=team_players.index, palette='magma')
plt.title('Top 15 Player Teams by Number of Player Records')
plt.xlabel('Number of Player Records')
plt.ylabel('Player Teams')
plt.show()
```

## 4. How does player weight vary across different teams?
**Business Question:** Are there noticeable differences in the physical profiles (weight) of players across the top AFL teams?

**Observations:**
- The median player weight is fairly consistent across most top teams, hovering around the 80-85kg mark.
- There are visible outliers in almost every team, representing exceptionally heavy or light players.

**Business Insight:** The consistency in weight distributions suggests a standard physical requirement across the league; however, analyzing outliers can help strength and conditioning coaches benchmark against league standards.

```
plt.figure(figsize=(14, 8))
top_teams = df['team'].value_counts().head(15).index
sns.boxplot(x='team', y='weight', data=df[df['team'].isin(top_teams)], palette='Set2')
plt.title('Player Weight Distribution Across Top 15 Teams')
plt.xlabel('Team')
plt.ylabel('Weight (kg)')
plt.xticks(rotation=45)
plt.show()
```

## 5. Which teams have the highest average fantasy points?
**Business Question:** Which teams generally produce the highest average fantasy points, indicating stronger individual statistical performances?

**Observations:**
- The top teams for average fantasy points show minimal variance, indicating a competitive balance in statistical output.
- Recent or traditionally high-scoring teams appear higher on the list.

**Business Insight:** Fantasy sports platforms can use this data to correctly price players from these high-performing teams, while clubs can use it to gauge overall team statistical dominance.

```
plt.figure(figsize=(12, 6))
team_fantasy = df.groupby('team')['avg_fantasy_points'].mean().sort_values(ascending=False).head(15)
sns.barplot(x=team_fantasy.values, y=team_fantasy.index, palette='coolwarm')
plt.title('Top 15 Teams by Average Fantasy Points')
plt.xlabel('Average Fantasy Points')
plt.ylabel('Team')
plt.show()
```

## 6. Who are the Top 15 goal scorers?
**Business Question:** Who are the most prolific goal scorers historically in the dataset?

**Observations:**
- The top 15 list is dominated by legendary forwards with massive career totals (e.g., Tony Lockett, Gordon Coventry).
- The gap between the #1 scorer and the 15th is significant, showing the exceptional nature of the very top players.

**Business Insight:** Historic top scorers drive high fan engagement. Marketing and media teams can leverage these legacy players for promotional content, documentaries, or special merchandise lines.

```
plt.figure(figsize=(12, 6))
top_scorers = df.groupby('player_full_name')['goals'].sum().sort_values(ascending=False).head(15)
sns.barplot(x=top_scorers.values, y=top_scorers.index, palette='rocket')
plt.title('Top 15 Goal Scorers')
plt.xlabel('Total Goals')
plt.ylabel('Player')
plt.show()
```

## 7. How many records are available for each season?
**Business Question:** How has the volume of recorded player statistics changed over the historical seasons?

**Observations:**
- There is a clear upward trend in the number of records available as time progresses.
- The recent decades show a plateau, representing the stabilization of the league size and standardized data collection.

**Business Insight:** Data from earlier years might be less reliable or sparser due to fewer records; analysts should weight recent data more heavily when building predictive models for modern games.

```
plt.figure(figsize=(14, 6))
season_counts = df['year'].value_counts().sort_index()
sns.lineplot(x=season_counts.index, y=season_counts.values, marker='o', color='green')
plt.title('Number of Records Available per Season')
plt.xlabel('Season (Year)')
plt.ylabel('Number of Records')
plt.grid(True)
plt.show()
```
