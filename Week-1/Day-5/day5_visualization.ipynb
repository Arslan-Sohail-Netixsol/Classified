# Day 5: Data Visualization with Pandas, Matplotlib, and Seaborn
This notebook visualizes the Titanic `tested.csv` dataset, fulfilling the 5 requested plots with explanations.


```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set visual style
sns.set_theme(style="whitegrid")

# Load and prep dataset (using same prep steps as Day 4)
file_path = 'D:/Python notes/archive/tested.csv'
df = pd.read_csv(file_path)

# Fill missing values to ensure smooth plotting
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Fare'] = df['Fare'].fillna(df['Fare'].median())

# Convert categorical data
df['Survived'] = df['Survived'].astype('category')
df['Pclass'] = df['Pclass'].astype('category')

print("Dataset ready for visualization!")
```

```
# Plot 1: Histogram of Age
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Passenger Ages', fontsize=14)
plt.xlabel('Age (Years)', fontsize=12)
plt.ylabel('Frequency (Number of Passengers)', fontsize=12)
plt.show()
```

**What this chart is telling us:**
This histogram displays the distribution of passenger ages on the Titanic. The bulk of the passengers were between 20 and 35 years old. The sharp peak around 27 represents our imputed median values (where we filled in the missing age data), showing how imputation can artificially spike the distribution at that single point. The smooth curved line (KDE) helps us see the general shape and trend of the ages regardless of the rigid bin boundaries.


```
# Plot 2: Bar Chart showing average Fare by Pclass
plt.figure(figsize=(8, 5))
avg_fare = df.groupby('Pclass', observed=False)['Fare'].mean().reset_index()
sns.barplot(x='Pclass', y='Fare', data=avg_fare, palette='viridis', hue='Pclass', legend=False)
plt.title('Average Ticket Fare by Passenger Class', fontsize=14)
plt.xlabel('Passenger Class (1 = 1st, 2 = 2nd, 3 = 3rd)', fontsize=12)
plt.ylabel('Average Fare ($)', fontsize=12)
plt.show()
```

**What this chart is telling us:**
This bar chart reveals the massive disparity in ticket pricing based on passenger class. First-class tickets cost significantly more on average (almost $100) compared to second and third-class tickets, which hover closely together around the $10-$20 mark. This strongly illustrates the economic divide among the passengers on board.


```
# Plot 3: Boxplot of Fare
plt.figure(figsize=(10, 4))
sns.boxplot(x=df['Fare'], color='lightgreen')
plt.title('Distribution and Outliers of Passenger Fares', fontsize=14)
plt.xlabel('Fare ($)', fontsize=12)
plt.show()
```

**What this chart is telling us:**
A boxplot is an excellent tool for spotting outliers. The green rectangular "box" represents the middle 50% of all ticket fares, showing that most people paid very little. However, the numerous black dots extending far to the right represent "outliers"—passengers who paid an exorbitant amount for their tickets, extending all the way past $500! These represent the elite, ultra-wealthy individuals in the highest-tier first-class suites.


```
# Plot 4: Scatter Plot of Age vs Fare, colored by Survival
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df, alpha=0.7, palette='coolwarm')
plt.title('Age vs Fare colored by Survival Status', fontsize=14)
plt.xlabel('Age (Years)', fontsize=12)
plt.ylabel('Fare ($)', fontsize=12)
plt.legend(title='Survived (0=No, 1=Yes)')
plt.show()
```

**What this chart is telling us:**
This scatter plot maps out each passenger based on two numeric variables: their Age (horizontal axis) and their Fare (vertical axis). The color tells us if they survived. We can see that the massive fare outliers (the dots high up on the y-axis) tend to be older adults. Visually, there doesn't seem to be a single isolated cluster of survivors, indicating that survival likely depended on a combination of factors (like class and gender) rather than just age or fare alone.


```
# Plot 5: Correlation Heatmap of Numeric Columns
plt.figure(figsize=(8, 6))
# Select only numeric columns for the correlation matrix
numeric_df = df.select_dtypes(include=['number'])
corr_matrix = numeric_df.corr()

sns.heatmap(corr_matrix, annot=True, cmap='RdBu', fmt='.2f', vmin=-1, vmax=1)
plt.title('Correlation Heatmap of Numeric Features', fontsize=14)
plt.show()
```

**What this chart is telling us:**
This heatmap displays the mathematical relationship (correlation) between all numeric columns. A value close to 1 means the two variables move up together, while a value close to -1 means as one goes up, the other goes down. 
For example, `Parch` (parents/children) and `SibSp` (siblings/spouses) have a somewhat strong positive correlation (0.31), which makes sense because large families tend to travel with both siblings and parents. `Age` and `Pclass` would likely have a negative correlation if we treated class numerically, as older people tend to be wealthier and thus in lower-numbered (better) classes.
