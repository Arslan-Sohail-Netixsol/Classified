```python
import pandas as pd

url = r"D:\Python notes\archive\tested.csv"

df = pd.read_csv(url)

# Question 1: What was the overall survival rate?
total_passengers = len(df)
survivors = len(df[df['Survived'] == 1])
survival_rate = (survivors / total_passengers) * 100
print(f"\n1. Overall survival rate: {survival_rate:.2f}% ({survivors} out of {total_passengers})")

# Question 2: How many passengers were in each class (1st, 2nd, 3rd)?
class_counts = df['Pclass'].value_counts().sort_index()
print("\n2. Passengers per class:")
for pclass, count in class_counts.items():
    print(f"   Class {pclass}: {count} passengers")
    
# Question 3: What was the survival rate for men vs. women?
women = df[df['Sex'] == 'female']
men = df[df['Sex'] == 'male']
women_survival_rate = (len(women[women['Survived'] == 1]) / len(women)) * 100
men_survival_rate = (len(men[men['Survived'] == 1]) / len(men)) * 100
print("\n3. Survival rate by gender:")
print(f"   Women: {women_survival_rate:.2f}%")
print(f"   Men: {men_survival_rate:.2f}%")

# Question 4: How many children (under 18) survived vs. died?
children = df[df['Age'] < 18]
children_survived = len(children[children['Survived'] == 1])
children_died = len(children[children['Survived'] == 0])
print(f"\n4. Children (under 18) survival:")
print(f"   Survived: {children_survived}")
print(f"   Died: {children_died}")


# Question 5: What was the average fare paid by passengers who survived vs. those who died?
survived_fare = df[df['Survived'] == 1]['Fare'].mean()
died_fare = df[df['Survived'] == 0]['Fare'].mean()
print("\n5. Average fare paid:")
print(f"   Survivors: ${survived_fare:.2f}")
print(f"   Non-survivors: ${died_fare:.2f}")

# Question 6: How many passengers embarked from each port?
embarked_counts = df['Embarked'].value_counts()
port_names = {'S': 'Southampton', 'C': 'Cherbourg', 'Q': 'Queenstown'}
print("\n6. Passengers embarked by port:")
for port_code, count in embarked_counts.items():
    print(f"   {port_names.get(port_code, port_code)}: {count}")
    
# Question 7: What was the survival rate of passengers in 1st class compared to 3rd class?
class1 = df[df['Pclass'] == 1]
class3 = df[df['Pclass'] == 3]
class1_survival = (len(class1[class1['Survived'] == 1]) / len(class1)) * 100
class3_survival = (len(class3[class3['Survived'] == 1]) / len(class3)) * 100
print("\n7. Survival rate by class (1st vs 3rd):")
print(f"   1st Class: {class1_survival:.2f}%")
print(f"   3rd Class: {class3_survival:.2f}%")

# Question 8: Did having family aboard (siblings/spouses) increase the chance of survival?
with_family = df[df['SibSp'] > 0]
alone = df[df['SibSp'] == 0]
family_survival = (len(with_family[with_family['Survived'] == 1]) / len(with_family)) * 100
alone_survival = (len(alone[alone['Survived'] == 1]) / len(alone)) * 100
print("\n8. Survival based on having siblings/spouses aboard:")
print(f"   With Family (SibSp > 0): {family_survival:.2f}% survival rate")
print(f"   Traveling Alone (SibSp = 0): {alone_survival:.2f}% survival rate")
```
